# Word2Vec, GloVe, FastText

Подгрузим набор данных с вопросами с Quora

```python
import numpy as np

with open("./data/quora.txt", encoding="utf-8") as file:
    data = list(file)

data[50]
```

Токенизация

```python
from nltk.tokenize import WordPunctTokenizer
tokenizer = WordPunctTokenizer()

print(tokenizer.tokenize(data[50]))
```

```python
tok_data = [tokenizer.tokenize(x.lower()) for x in data]
```

```python
tok_data[50]
```

Можно построить w2v и fasttext на наших данных

```python
from gensim.models import Word2Vec
model_w2v = Word2Vec(tok_data, 
                     vector_size=32,
                     min_count=5,
                     window=5).wv
```

```python
from gensim.models import FastText
model_ft = FastText(tok_data, 
                    vector_size=32,
                    min_count=5,
                    window=5).wv
```

```python
model_w2v.get_vector('auto')
```

```python
model_ft.get_vector('auto')
```

```python
model_w2v.most_similar('auto')
```

```python
model_ft.most_similar('auto')
```

предобученные модели

```python
import gensim.downloader as api
api.info()
```

```python
model = api.load('glove-twitter-100')
```

```python
model.most_similar(positive=["king", "woman"], negative=["man"])
```

```python
model.most_similar(positive=["coder", "money"], negative=["work"])
```

Визуализация

1000 наиболее частых слов

```python
words = model.index_to_key[:1000] 

print(words[::100])
```

```python
word_vectors = [model.get_vector(word) for word in words]
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

word_vectors_pca = pca.fit_transform(word_vectors)
```

```python
import bokeh.models as bm, bokeh.plotting as pl
from bokeh.io import output_notebook
output_notebook()

def draw_vectors(x, y, radius=10, alpha=0.25, color='blue',
                 width=600, height=400, show=True, **kwargs):
    """ draws an interactive plot for data points with auxilirary info on hover """
    if isinstance(color, str): color = [color] * len(x)
    data_source = bm.ColumnDataSource({ 'x' : x, 'y' : y, 'color': color, **kwargs })

    fig = pl.figure(active_scroll='wheel_zoom', width=width, height=height)
    fig.scatter('x', 'y', size=radius, color='color', alpha=alpha, source=data_source)

    fig.add_tools(bm.HoverTool(tooltips=[(key, "@" + key) for key in kwargs.keys()]))
    if show: pl.show(fig)
    return fig
```

```python
draw_vectors(word_vectors_pca[:, 0], word_vectors_pca[:, 1], token=words)
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2)

word_vectors_tsne = tsne.fit_transform(np.array(word_vectors))
```

```python
draw_vectors(word_vectors_tsne[:, 0], word_vectors_tsne[:, 1], token=words)
```
